# Tokenization Using NLTK

**Tokenization** is the process of breaking down text into smaller units, called tokens. These tokens can be words, sentences, or subwords, depending on the task. In **Natural Language Processing (NLP)**, tokenization is a crucial preprocessing step for tasks like text classification, sentiment analysis, and machine translation.

## Features of Tokenization in NLTK
- **Word Tokenization**: Splits text into individual words.
- **Sentence Tokenization**: Splits text into sentences.
- Prebuilt tokenizers handle complexities like punctuation, abbreviations, and contractions.

## Examples of Tokenization with NLTK

### 1. Word Tokenization
```python
from nltk.tokenize import word_tokenize

text = "NLTK is a leading library for NLP."
tokens = word_tokenize(text)
print(tokens)
# Output: ['NLTK', 'is', 'a', 'leading', 'library', 'for', 'NLP', '.']
```

### 2. Sentence Tokenization
```python
from nltk.tokenize import sent_tokenize

text = "NLTK is a leading library for NLP. It supports tokenization!"
sentences = sent_tokenize(text)
print(sentences)
# Output: ['NLTK is a leading library for NLP.', 'It supports tokenization!']
```

## Custom Tokenization
For domain-specific text, NLTK provides tools like `RegexpTokenizer` and `TweetTokenizer`:
- **RegexpTokenizer**: Uses regular expressions for custom patterns.
- **TweetTokenizer**: Designed for tokenizing social media text, including hashtags and emojis.

### Example: RegexpTokenizer
```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize("Tokenize this sentence, please!")
print(tokens)
# Output: ['Tokenize', 'this', 'sentence', 'please']
```

## Applications of Tokenization
- **Text Preprocessing**: Tokenization is the first step in converting raw text into numerical formats for machine learning.
- **Sentiment Analysis**: Helps break down text to analyze opinions.
- **Information Retrieval**: Splits queries and documents into words for matching.

## Limitations
- May not handle all edge cases (e.g., abbreviations or domain-specific jargon).
- Language-specific nuances require specialized tokenization strategies.

### **Definition of Stemming in Machine Learning**

**Stemming** is a text preprocessing technique used in **Natural Language Processing (NLP)** to reduce words to their root or base form (called the "stem"). The primary goal of stemming is to remove suffixes (and sometimes prefixes) from words, ensuring that words with similar meanings are treated as a single term during text analysis.

Unlike **lemmatization**, stemming does not necessarily produce a linguistically accurate word but a shortened version or stem.

---

### The Problem Statement
- ***Word Variability in Text Data:*** Natural language is diverse, and the same concept can be expressed in multiple word forms due to inflections, derivations, or conjugations. For example:

Words like "run," "running," "runner," "ran" represent the same underlying concept (action of running) but appear as different tokens to the model.
Without stemming, these variations increase:
Vocabulary size.
Redundancy in features.
Computational complexity.
- ***Challenges in Text-Based ML Models:***
- ***Sparse Feature Space:***

Words in different forms are treated as separate features, leading to sparsity in the dataset.
Example: A model might consider "run" and "running" as two distinct features even though they are semantically related.
Reduced Generalization:

Models may fail to capture relationships between word variations, affecting predictions.
Example: In sentiment analysis, the model might not associate "happiness" and "happily" with the same positive sentiment.
- ***Inefficient Search and Retrieval:***

Information retrieval systems might miss relevant results because of word mismatches.

---
### **How Stemming Works**

Stemming is rule-based and often uses algorithms like:
- **Porter Stemmer**
- **Snowball Stemmer**
- **Lancaster Stemmer**
- **Regexp stemmer**
- **ISRI Stemmer**

These algorithms apply predefined rules to chop off word endings such as `-ing`, `-ed`, `-s`, etc.



---

### **Examples of Stemming**

#### Example 1: Basic Words
| **Word**      | **Stemmed Form** |
|----------------|------------------|
| **running**   | run              |
| **runner**    | run              |
| **ran**       | ran              |
| **easily**    | easi             |
| **studies**   | studi            |

#### Example 2: Similar Stems for Variations
| **Word**         | **Stemmed Form** |
|-------------------|------------------|
| **connected**     | connect          |
| **connection**    | connect          |
| **connections**   | connect          |

#### Example 3: Aggressive Stemming (Lancaster Stemmer)
| **Word**      | **Stemmed Form** |
|---------------|------------------|
| **troubling** | troubl           |
| **troubled**  | troubl           |
| **troubles**  | troubl           |

---

### **Applications in Machine Learning**
1. **Information Retrieval**:
   - Stemming helps search engines treat variations of a word (e.g., "run" and "running") as equivalent, improving search results.
   
2. **Text Classification**:
   - Reduces dimensionality in feature space by consolidating word variations.
   
3. **Topic Modeling**:
   - Groups related terms, simplifying topic extraction.

---

### **Limitations of Stemming**
- Stemming may over-stem or under-stem words:
  - **Over-stemming**: Cutting too much (e.g., "universe"  "univers").
  - **Under-stemming**: Not cutting enough (e.g., "running"  "running").
- It doesn't always produce a valid dictionary word.
- Can lead to ambiguity, as stems like "connect" may combine unrelated terms.
- Stemming <span style="color: red;">*lacks context*</span>  leading to innaccuracies

---
### **When to use what type of stemmer:**
- **Porter**: need standard, well-recognized stemmer for English with reasonable accuracy.
- **Snowball**: need better linguistic accuracy, cross-language support than Porter.
- **Lancaster**: If speed is crucial, and you can tolerate some over-stemming.

````python
## Stemmers comparison
#import time


from nltk.stem import PorterStemmer, SnowballStemmer,LancasterStemmer
porter_list, snowball_list, lancaster_list = [],[],[]

# Record the start time
#start_time = time.time()

porter_stem=PorterStemmer()
snowball_stem=SnowballStemmer(language='english')
lancaster_stem=LancasterStemmer()

#stemming.stem(words_for_play)
for word in words_for_play:
    porter_list.append(porter_stem.stem(word))
    snowball_list.append(snowball_stem.stem(word))
    lancaster_list.append(lancaster_stem.stem(word))


# Creating a DataFrame
df = pd.DataFrame({
    "Original": words_for_play,
    "Porter": porter_list,
    "Snowball": snowball_list,
    "Lancaster": lancaster_list
})

df.head(10)


#end_time = time.time()
# Calculate the elapsed time in microseconds
#elapsed_time_microseconds = (end_time - start_time) * 1_000_000  # Convert to microseconds
#print(f"Elapsed time: {elapsed_time_microseconds} microseconds")
````

# Lemmatization 

Lemmatization is a process in Natural Language Processing (NLP) that reduces words to their **base or root form** <span style="color: red;"> not the root stem</span>. Unlike stemming, which cuts off prefixes or suffixes, lemmatization takes into account the **meaning** of the word and uses a vocabulary and morphological analysis to transform the word into its correct base form, called a **lemma**.

### Key Points:
- **Goal**: To convert words like *better* into *good* and *running* into *run* based on their meanings.
- **More Accurate Than Stemming**: Lemmatization ensures that the word is transformed into a valid word in the language, unlike stemming, which may produce non-existent words.
- **Requires POS (Part of Speech) Tagging**: To get the correct lemma, lemmatization often requires knowing whether the word is a noun, verb, adjective, etc.

### Example:
- **Word**: *"running"*
  - **Lemmatized word**: *"run"*
- **Word**: *"better"*
  - **Lemmatized word**: *"good"*

### Tools for Lemmatization:
In Python, the **NLTK** and **spaCy** libraries provide built-in functions for lemmatization:
- **NLTK**: Uses WordNet Lemmatizer.
- **spaCy**: Has an efficient lemmatizer that works with pre-trained models.

### Why Use Lemmatization in NLP?
- It reduces different word forms to a common base, making text processing more efficient.
- Helps improve the accuracy of tasks like text classification, sentiment analysis, and machine translation by handling word variations in a more meaningful way.

Lemmatization helps ensure that variations of a word are treated as the same word, which can enhance the performance of NLP models.


# WordNetLemmatizer: A Quick Introduction

The **WordNetLemmatizer** is a lemmatization tool in the **NLTK (Natural Language Toolkit)** library that uses the **WordNet** lexical database to transform words into their base or root form (lemma). Unlike stemming, which cuts off word endings, lemmatization considers the word's meaning and context, ensuring the result is a valid word.

### Key Features:
- **Uses WordNet**: A large database that groups words into **synsets** (sets of synonyms).
- **Context-Aware**: Requires **part of speech (POS)** tags to accurately lemmatize words based on their meaning.
- **Accurate**: Returns valid base forms (e.g., "better"  "good" as an adjective, "running"  "run").

### Example:

```python
from nltk.stem import WordNetLemmatizer

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatizing words
print(lemmatizer.lemmatize("running", pos="v"))  # Outputs: "run"
print(lemmatizer.lemmatize("better", pos="a"))   # Outputs: "good"
```
---

### When to Use WordNetLemmatizer:

- ***Text Normalization***: To reduce words to their base form for consistency.
- ***Improving NLP Models***: It helps in handling word variations more accurately, improving model performance in tasks like classification, translation, and information retrieval

### 

### Part of Speech (POS) in NLP

**Part of Speech (POS) tagging** is the process of identifying the category of each word in a sentence based on its function. POS tags help machines understand the grammatical structure of a sentence, enabling better language processing for tasks such as **text classification**, **sentiment analysis**, and **machine translation**.

Common POS categories include:

1. **Nouns (NN)**: Represent people, places, things, or ideas (e.g., *dog*, *city*).
2. **Verbs (VB)**: Express actions or states (e.g., *run*, *be*).
3. **Adjectives (JJ)**: Modify nouns (e.g., *happy*, *quick*).
4. **Adverbs (RB)**: Modify verbs, adjectives, or other adverbs (e.g., *quickly*, *very*).
5. **Pronouns (PRP)**: Replace nouns (e.g., *he*, *she*).

### Importance in NLP:
- Helps in **understanding sentence structure** and relationships between words.
- Facilitates **feature extraction** for text classification tasks.
- Enhances **machine translation** by understanding syntax.

### Applications:
- **Named Entity Recognition (NER)**: Identifying entities like names or locations.
- **Information Extraction**: Extracting relevant information from text.
- **Text Preprocessing**: Essential for lemmatization and text cleaning.

POS tagging is a crucial step in various NLP workflows, providing insights into the meaning and structure of text data.
